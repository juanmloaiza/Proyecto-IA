**Extracción y carga de los archivos desde una carpeta contenida en Google Drive:**
Esta sección conecta el Google Drive con Colab para acceder a los archivos contenidos en dicha carpeta, específicamente al archivo train.csv que contiene los datos de estudiantes y sus resultados en las pruebas. Luego se usa la librería pandas para leer ese archivo y almacenarlo en una estructura de tabla que facilita el análisis de datos y finalmente muestra las primeras cinco filas del archivo para verificar que se cargó correctamente.

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

file_path = '/content/drive/MyDrive/Proyecto IA/Pruebas_Saber_Pro/train.csv'

df = pd.read_csv(file_path)

df.head()
```

**Exploración inicial:**
Se realizó una exploración básica de los datos cargados desde el archivo train.csv, que se enfoca en tres columnas específicas.
Primero se analizó la columna RENDIMIENTO_GLOBAL, que clasifica a los estudiantes según su desempeño en las pruebas (alto, medio-alto, medio-bajo y bajo). Luego se examinó la columna ESTU_PRGM_DEPARTAMENTO que indica el departamento del país donde estudia cada persona y finalmente se revisó la columna FAMI_TIENECOMPUTADOR, la cual señala si el estudiante cuenta o no con computador en su hogar.

```python
#Clasificación del estudiante según rendimiento
conteo_rendimiento = df['RENDIMIENTO_GLOBAL'].value_counts()
print("Cantidad de estudiantes por nivel de rendimiento:")
print(conteo_rendimiento)

#Número de estudiantes relacionado a cada departamento
if 'ESTU_PRGM_DEPARTAMENTO' in df.columns:
    conteo_departamento = df['ESTU_PRGM_DEPARTAMENTO'].value_counts()
    print("\nCantidad de estudiantes por departamento del programa:")
    print(conteo_departamento)
else:
    print("\nNo se encontró la columna 'ESTU_PRGM_DEPARTAMENTO'.")

#Caracterización de estudiantes con computador o no
if 'FAMI_TIENECOMPUTADOR' in df.columns:
    conteo_computador = df['FAMI_TIENECOMPUTADOR'].value_counts()
    print("\nCantidad de estudiantes según si tienen computador en casa:")
    print(conteo_computador)
else:
    print("\nNo se encontró la columna 'FAMI_TIENECOMPUTADOR'.")


```
